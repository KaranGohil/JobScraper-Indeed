## Exploratory Data Analysis (EDA)

In EDA, I'm just exploring my scaper data and I will also be calculating the count frequency which will get me the number of times a word occurs and also tfidf.

Term Frequency-Inverse Document Frequency (tfidf) is used to give weight to the words which are less occurring in the document, these weights help figure out how important a word is in a document.

I will be processing all the job descriptions to remove all the stop words and punctuation. The description will also go through the Porter Stemming algorithm to decrease the overall noise of the description. This helps in finding the relevant info by removing unnecessary words.

```python
# Preprocessing
fsd['Description'] = fsd['Description'].apply(lambda x: mf.preprocessing(x))
sd['Description'] = sd['Description'].apply(lambda x: mf.preprocessing(x))
da['Description'] = da['Description'].apply(lambda x: mf.preprocessing(x))
ds['Description'] = ds['Description'].apply(lambda x: mf.preprocessing(x))
fed['Description'] = fed['Description'].apply(lambda x: mf.preprocessing(x))
```

As we can see there is too much noise while calculating the count Frequency and tfidf. Therefore, I have come up with a naive solution. I'm going to create a general list of technical skills and grab those technical skills' count Frequency and tfidf.

```python
print("----------------------Software Developer------------------")
hs_cf_sd = mf.searchHardSkills(cf_sd).transpose().reset_index().rename(columns={'index': 'technical skills', 0: 'count'})
display(hs_cf_sd)
print("----------------------Data Analyst------------------")
hs_cf_da = mf.searchHardSkills(cf_da).transpose().reset_index().rename(columns={'index': 'technical skills', 0: 'count'})
display(hs_cf_da)
print("----------------------Data Scientist------------------")
hs_cf_ds = mf.searchHardSkills(cf_ds).transpose().reset_index().rename(columns={'index': 'technical skills', 0: 'count'})
display(hs_cf_ds)
print("----------------------Full Stack Developer------------------")
hs_cf_fsd = mf.searchHardSkills(cf_fsd).transpose().reset_index().rename(columns={'index': 'technical skills', 0: 'count'})
display(hs_cf_fsd)
print("----------------------Front End Developer------------------")
hs_cf_fed = mf.searchHardSkills(cf_fed).transpose().reset_index().rename(columns={'index': 'technical skills', 0: 'count'})
display(hs_cf_fed)
```

```python
sns.set(rc = {'figure.figsize':(15,8)})
sns.barplot(x= 'technical skills', y='count', data=hs_cf_sd).set_title("Software Developer")
plt.xticks(rotation=45)
```

# Analysis

As we know the all desired technical skills for each job title, we will put each technical skill and the count into a vector form before that will make sure all the job title has the same technical skills. If one of the technical skills exists for one job title and not for the other then that technical skill will have a count of 0. This will make sure the result is accurate and fair. In the end, we will calculate the cosine similarity between all the job titles.

### Hypothesis

1. As Data Analyst and Data Scientist are from the same field, they will have higher similarity score.
2. Same goes for Full Stack Developer and Front End Developer.
3. I also think the similarity between Software Developer and Full Stack Developer will be high.
4. While the data field jobs will have a low similarity with Software Developer, Full Stack Developer and Front End Developer

```python
print(f"1. The cosine similarity between Data Analyst and Data Scientist is {mf.getCosineSim(hs_cf_da, hs_cf_ds)}")
print(f"2. The cosine similarity between Full Stack Developer and Front End Developer is {mf.getCosineSim(hs_cf_fsd, hs_cf_fed)}")
print(f"3. The cosine similarity between Software Developer and Full Stack Developer is {mf.getCosineSim(hs_cf_sd, hs_cf_fsd)}")
print(f"4(a). The cosine similarity between Data Analyst and Software Developer is {mf.getCosineSim(hs_cf_da, hs_cf_sd)}")
print(f"4(b). The cosine similarity between Data Analyst and Full Stack Developer is {mf.getCosineSim(hs_cf_da, hs_cf_fsd)}")
print(f"4(c). The cosine similarity between Data Analyst and Front End Developer is {mf.getCosineSim(hs_cf_da, hs_cf_fed)}")
print(f"4(d). The cosine similarity between Data Scientist and Software Developer is {mf.getCosineSim(hs_cf_ds, hs_cf_sd)}")
print(f"4(e). The cosine similarity between Data Scientist and Full Stack Developer is {mf.getCosineSim(hs_cf_ds, hs_cf_fsd)}")
print(f"4(f). The cosine similarity between Data Scientist and Front End Developer is {mf.getCosineSim(hs_cf_ds, hs_cf_fed)}")
```

# Conclusion 

- As you can see my first hypothesis was correct, Data Analyst and Data Scientist has a high cosine similarity. Therefore, they have similar technical skill which are desired from the employers/job posters.
- My second hypothesis was also correct, Full Stack Developer and Front End Developer has a high cosine similarity.
- My third hypothesis was also correct, Software Developer and Full Stack Developer has a high cosine similarity.
